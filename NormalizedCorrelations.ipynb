## Correlation Sum of States

```python
labels = ["source", "target", "corr"]
data = pd.read_csv("res/correlations/normalize/correlations/corr_sum_state/part-00000", names=labels)
data = data.pivot("source", "target", "corr")
fig, ax = plt.subplots(figsize=(20, 20))
sns.heatmap(data)
```

## Correlation Sum of States Opioids

```python
data = pd.read_csv("res/correlations/normalize/correlations/corr_sum_state_opioids/part-00000", names=labels)
data = data.pivot("source", "target", "corr")
fig, ax = plt.subplots(figsize=(20, 20))
sns.heatmap(data)
```
