# Data Modeling

```python
#Python Sum State Values Per Year
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

#Read in data of single year
data_2000 = pd.read_csv('/Users/andrehochmuth/documents/cs435/project/database/State_Drug_Utilization_Data_2000.csv',low_memory=False)


#Remove all Rows that Have Condition:  Suppression Used == True to rid of suppressed data which is not reliable.
data_2000 = data_2000.loc[data_2000['Suppression Used'] == False]
#Remove all Rows that do not have a state recorded
data_2000 = data_2000.loc[data_2000['State'] != 'XX']

#Remove Unnecessary Information (Columns)
national_2000 = data_2000.drop(['Utilization Type', 'Labeler Code', 'Product Code', 'Package Size', 'Quarter','Product Name', 'Number of Prescriptions', 'Quarter Begin', 'Quarter Begin Date', 'Latitude', 'Longitude', 'Location', 'Total Amount Reimbursed', 'Year', 'Suppression Used', 'Medicaid Amount Reimbursed', 'Non Medicaid Amount Reimbursed'],axis=1)
national_2000.columns = ['State','Units Reimbursed','ndc']

#Sum each ndc's units reimbursed
national_2000 = national_2000.groupby('State')['Units Reimbursed'].sum()
national_2000.columns = ['State','Units Reimbursed']
```
