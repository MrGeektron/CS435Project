## Modeling State Trends

```python
#Python Sum State Values Per Year
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

#Read in data of single year
data_2000 = pd.read_csv('/Users/andrehochmuth/documents/cs435/project/database/State_Drug_Utilization_Data_2015.csv',low_memory=False)
poop = pd.read_csv('/Users/andrehochmuth/documents/cs435/project/database/State_Drug_Utilization_Data_2000.csv',low_memory=False)

#Read State Population Data
Populations = pd.read_csv('/Users/andrehochmuth/documents/cs435/project/StatePopulation.csv')

#Remove all Rows that Have Condition:  Suppression Used == True to rid of suppressed data which is not reliable.
data_2000 = data_2000.loc[data_2000['Suppression Used'] == False]
poop = poop.loc[poop['Suppression Used'] == False]

#Remove all Rows that do not have a state recorded
data_2000 = data_2000.loc[data_2000['State'] != 'XX']
poop = poop.loc[poop['State'] != 'XX']


#Remove Unnecessary Information (Columns)
national_2000 = data_2000.drop(['Utilization Type', 'Labeler Code', 'Product Code', 'Package Size', 'Quarter','Product Name', 'Number of Prescriptions', 'Quarter Begin', 'Quarter Begin Date', 'Latitude', 'Longitude', 'Total Amount Reimbursed', 'Year', 'Suppression Used', 'Medicaid Amount Reimbursed', 'Non Medicaid Amount Reimbursed','Location'],axis=1)
national_2000.columns = ['State','Units Reimbursed','ndc']
poop = poop.drop(['Utilization Type', 'Labeler Code', 'Product Code', 'Package Size', 'Quarter','Product Name', 'Number of Prescriptions', 'Quarter Begin', 'Quarter Begin Date', 'Latitude', 'Longitude', 'Total Amount Reimbursed', 'Year', 'Suppression Used', 'Medicaid Amount Reimbursed', 'Non Medicaid Amount Reimbursed','Location'],axis=1)
poop.columns = ['State','Units Reimbursed','ndc']

#Sum each ndc's units reimbursed
national_2000 = national_2000.groupby('State')['Units Reimbursed'].sum().reset_index()
national_2000.columns = ['state','Units Reimbursed']
national_2000 = national_2000.sort_values(by='Units Reimbursed')

poop = poop.groupby('State')['Units Reimbursed'].sum().reset_index()
poop.columns = ['state','Units Reimbursed']
poop = poop.sort_values(by='Units Reimbursed')
```

```python
#ChloroPleth Map
import plotly
plotly.offline.init_notebook_mode(connected=True)
import plotly.offline as py
import pandas as pd
#plotly.tools.set_credentials_file(username='AndreHochmuth', api_key='')
df = national_2000

for col in df.columns:
    df[col] = df[col].astype(str)

scl = [[0.0, 'rgb(242,240,247)'],\
            [0.1, 'rgb(117,107,177)'],[1, 'rgb(84,39,143)']]

df['text'] = df['state'] + '<br>' +\
    'Total Reimbursement Amount: '+ df['Units Reimbursed'] 

data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = df['state'],
        z = df['Units Reimbursed'].astype(float)-poop['Units Reimbursed'].astype(float)/(Populations['2015'].astype(float)-Populations['2000'].astype(float)),
        locationmode = 'USA-states',
        text = df['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            )
        ),
        colorbar = dict(
            title = "Units Reimbursed"
        )
    ) ]

layout = dict(
        title = 'Ratio of Increase in Total Drugs Reimbursed by Population Increase From 2000 to 2015',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)',
        ),
    )

fig = dict(data=data, layout=layout)

url = py.plot(fig, filename='Opioid_Map.html')
```
